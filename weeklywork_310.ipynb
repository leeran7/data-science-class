```python
import pandas as pd
import numpy as mp
import matplotlib.pyplot as plt
df = pd.read_csv("austin_weather.csv");
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.dtypes
```

```python
# Sort values by date

df.sort_values(by='Date', inplace=True)
df
```

```python
# Take all object types in the columns EXCEPT events

cols = df.columns[df.dtypes.eq('object')][:-1]
cols
```

```python
#change all object types that need to be numeric

df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
df.dtypes
```
