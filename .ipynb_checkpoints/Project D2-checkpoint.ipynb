# Lets use Linear Regression to test the data set
## Before we run Linear regression, lets clean the data of any strings

```python
##drop string columns
df.drop(['name', 'host_name', 'last_review', 'neighbourhood', 'neighbourhood_group', 'room_type', 'reviews_per_month'], inplace=True, axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('price', axis=1), df['price'], test_size = .2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
y_pred_test = lm.predict(X_test)
y_pred_train = lm.predict(X_train)
```

```python
residuals = abs(y_test - y_pred_test)
indexes = list(residuals.nlargest(10).index)
worst = y_test[indexes]
test = [y_test.index.get_loc(x) for x in indexes]
```

```python
sns.scatterplot(y_test, y_pred_test)
sns.scatterplot(worst, y_pred_test[test], color='red')
```
