```python
mall = pd.read_csv("Mall_Customers(1).csv")
mall.head()
```

```python
plt.scatter(mall['Age'], mall['Spending Score (1-100)'])
plt.title("Customer spending score to Age")
```

```python
plt.hist(mall['Age'], bins=30)
```

```python
plt.hist(mall['Annual Income (k$)'])
```

```python
plt.hist(mall[mall['Gender'] == 'Female']['Spending Score (1-100)'], alpha=.5);
plt.hist(mall[mall['Gender'] == 'Male']['Spending Score (1-100)'], alpha=.7);
plt.title("Female to Male spending score")
plt.legend(['Female', 'Male'])
```

```python
sns.pairplot(mall)
```

```python
sns.histplot(mall['Annual Income (k$)'], kde=True);
```

```python
sns.displot(mall, x="Annual Income (k$)", kde=True)
```

```python
sns.displot(mall, x="Annual Income (k$)", kind="kde")
```

```python
sns.displot(mall, x="Annual Income (k$)", rug=True)
```
