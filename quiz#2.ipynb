```python
np.random.seed(42)

mu = 175 # mean height in centimeters

sigma = 6 # standard deviation

sample = 100 # sample size

h = np.random.normal(mu, sigma, sample) # get 100 random datapoints using the mean mu and std deviations sigman

w = (h - 100) *np.random.uniform(0.75, 1.25, 100)

stats = pd.DataFrame({"heightcm": h,

        "weightkg": w}) #stats is a dataframe with 2 columns
```

```python
stats.head()
```

```python
plt.scatter(x='heightcm', y='weightkg', data=stats, s=100, alpha=.3, marker='bo')
```
